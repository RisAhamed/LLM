```python
import os
import requests
from bs4 import BeautifulSoup

def save_text_from_url(url, directory):
    # Fetch the content from the URL
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract text from the body of the page
        text = soup.get_text(separator='\n', strip=True)
        
        # Ensure the directory exists
        if not os.path.exists(directory):
            os.makedirs(directory)
        
        # Define the file path
        file_path = os.path.join(directory, 'extracted_text.txt')
        
        # Save the extracted text to a file
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(text)
        
        print(f'Text saved to {file_path}')
    else:
        print(f'Failed to retrieve the page. Status code: {response.status_code}')

# Example usage
url = 'https://en.wikisource.org/wiki/The_Verdict'
directory = 'Verdict.txt'
save_text_from_url(url, directory)
```

```python
from idna import valid_contextj


class SimpleTokenizerV1:
    def __init__(self,vacob):
        self.str_to_int = vacob
        self.int_to_str = { i:s for s,i in vacob.items()}

    def encode(self,text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [
        item.strip() for item in preprocessed if item.strip()
        ]
        ids =[ self.str_to_int[s] for s in preprocessed]
        return ids
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
```

```python
class SimpleTokenizerV2:

    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}
    
    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [
            item if item in self.str_to_int 
            else "<|unk|>" for item in preprocessed
        ]

        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
        
    def decode(self, ids):
        import re
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
from torch.utils.data import Dataset,DataLoader
import torch
import tiktoken
# from math import range,len
class GPTDatasetV1(Dataset):
    def __init__(self,text,tokenizer,max_length,stride):
        self.input_ids = []
        self.target_ids = []

        # tokenizer = tiktoken.get_encoding("gpt2")
        token_ids = tokenizer.encode(text,allowed_special={"|endoftext|"})
        for i in range(0,len(token_ids)- max_length,stride):
            input_chuk =token_ids[i:i + max_length]
            target_chunk = token_ids[i+1 : i+max_length+1]
            self.input_ids.append(torch.tensor(input_chuk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self):
        return len(self.input_ids)
    def __getitem__(self, index):
        return self.input_ids[index],self.target_ids[index]
    
```

```python
from torch.utils.data import DataLoader
def create_dataloader_v1(txt, tokenizer, batch_size, max_length=256, stride=128, shuffle=True, drop_last=True, num_workers=0):
    dataset = GPTDatasetV1(text=txt, tokenizer=tokenizer, max_length=max_length, stride=stride)
    dataloader = DataLoader(dataset=dataset, batch_size=batch_size, shuffle=shuffle, drop_last=drop_last, num_workers=num_workers)
    return dataloader
```

## TORCH MANULA  EMBEDDINGS

```python
import torch
import torch.nn as nn
import numpy as np

# Sample vocabulary and embeddings for demonstration
vocab = {0: "hello", 1: "world", 2: "foo", 3: "bar", 4: "baz"}
vocab_to_index = {v: k for k, v in vocab.items()}

# Create an embedding layer
embedding_dim = 8
embedding = nn.Embedding(num_embeddings=len(vocab), embedding_dim=embedding_dim)

# Example embedding matrix (initialized randomly for demonstration)
embedding_matrix = embedding.weight.data

# Function to convert embedding back to word using index lookup
def embedding_to_word(embedding_vector, embedding_matrix, vocab):
    # Compute similarity between the input vector and all vectors in the embedding matrix
    similarity = torch.matmul(embedding_matrix, embedding_vector)

    closest_index = torch.argmax(similarity).item()

    return vocab[closest_index]

# Get a sample embedding for "hello"
input_index = vocab_to_index["hello"]
embedding_vector = embedding(torch.tensor(input_index))

# Convert back to word
reconstructed_word = embedding_to_word(embedding_vector, embedding_matrix, vocab)
print(f"Original word:",embedding_vector ,"Reconstructed word: '{reconstructed_word}'")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def approximate_embedding_to_word(embedding_vector, embedding_matrix, vocab):
    # Convert tensors to numpy arrays
    embedding_vector = embedding_vector.detach().numpy()
    embedding_matrix = embedding_matrix.detach().numpy()

    # Compute cosine similarity
    similarities = cosine_similarity(embedding_matrix, embedding_vector.reshape(1, -1))
    closest_index = np.argmax(similarities)

    return similarities,closest_index,vocab[closest_index]

# Example usage
similarities,closest_index,approx_reconstructed_word = approximate_embedding_to_word(embedding_vector, embedding_matrix, vocab)
print(f"Approximate Reconstructed word: '{approx_reconstructed_word}'")
print(similarities,closest_index)
```
